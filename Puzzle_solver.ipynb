```python


import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


# Define the board (19x18 grid)
board_str = [
"XXXXXXXXXXX00000000",
"XXXXXXXXXXX00000000",
"XXXXXXXXXXX00000000",
"XXXXXXXXX0000000000",
"XXXXXX0000000000000",
"XXXXX00000000000000",
"XXXXX00000000000000",
"XXXXX00000000000000",
"XXX0000000000000000",
"XXX000000000000000X",
"XXX0000000000000XXX",
"X000000000000000XXX",
"00000000000000XXXXX",
"0000000000000XXXXXX",
"0000000000XXXXXXXXX",
"0000000000XXXXXXXXX",
"0000000000XXXXXXXXX",
"000000000XXXXXXXXXX",
"000000000XXXXXXXXXX"
]

BOARD_COLS = 19;
BOARD_ROWS = 19;

# Convert the board to a NumPy array (0 = free, -1 = blocked)
board = np.array([[0 if cell == '0' else -1 for cell in row] for row in board_str])

# Hardcoded paths for each piece
piece_paths = {
    1: [
"rrrruuurrruuuurrurruuurrrrr",
"uuuurrruuurrrruuruurrruuuuu",
"rrrrruuurrurruuuurrruuurrrr",
"uuuuurrruuruurrrruuurrruuuu"
    ],
    2: [
"rrruuuruurrrruuruuurrruurrr",
"uuurrrurruuuurrurrruuurruuu",
"rrruurrruuuruurrrruuruuurrr",
"uuurruuurrrurruuuurrurrruuu"
    ],
    3: [
"rruurrruurrrurruuuruuurrurrr",
"uurruuurruuuruurrrurrruuruuu",
"rrrurruuuruuurrurrruurrruurr",
"uuuruurrrurrruuruuurruuurruu"
    ],
    4: [
"rrruuuurrruurrurruruuurrurr",
"uuurrrruuurruuruururrruuruu",
"rrurruuururrurruurrruuuurrr",
"uuruurrruruuruurruuurrrruuu"

    ]
}

# Convert paths to coordinates
def path_to_coords(path):
    x, y = 0, 0
    coords = [(0, 0)]  # Starting point
    for move in path:
        if move == 'r':
            y += 1
        elif move == 'u':
            x -= 1
        elif move == 'l':
            y -= 1
        elif move == 'd':
            x += 1
        coords.append((x, y))
    return coords

# Rotate coordinates 90 degrees
def rotate_coords(coords):
    return [(y, -x) for (x, y) in coords]

# Check if a piece can be placed at a specific position
def can_place_piece(board, coords, x, y):
    for (dx, dy) in coords:
        nx, ny = x + dx, y + dy
        if nx < 0 or nx >= board.shape[0] or ny < 0 or ny >= board.shape[1]:
            return False  # Out of bounds
        if board[nx][ny] != 0:
            return False  # Overlaps with blocked space or another piece
    return True

# Place a piece on the board
def place_piece(board, coords, x, y, piece_id):
    new_board = np.copy(board)
    for (dx, dy) in coords:
        nx, ny = x + dx, y + dy
        new_board[nx][ny] = piece_id
    return new_board

# Remove a piece from the board
def remove_piece(board, coords, x, y):
    new_board = np.copy(board)
    for (dx, dy) in coords:
        nx, ny = x + dx, y + dy
        new_board[nx][ny] = 0
    return new_board

# Visualize the piece on the board
def visualize_piece(board, coords, x, y, piece_id):
    new_board = np.copy(board)
    for (dx, dy) in coords:
        nx, ny = x + dx, y + dy
        if 0 <= nx < board.shape[0] and 0 <= ny < board.shape[1]:
            new_board[nx][ny] = piece_id
    print("Visualizing piece:")
    print(new_board)

# Visualize the board using Matplotlib
def draw_board(board):
    """Visualize the board using Matplotlib with different colors for different pieces."""

    # Define a custom color map for different pieces
    cmap = mcolors.ListedColormap([
        'tan',  # 0 - Empty space
        'white',    # 1 - Piece 1
        'brown',   # 2 - Piece 2
        'chocolate',  # 3 - Piece 3
        'sandybrown', # 4 - Piece 4
        'goldenrod'    # -1 - Blocked area
    ])

    # Define boundaries for colors
    bounds = [-1.5, -0.5, 0.5, 1.5, 2.5, 3.5, 4.5]
    norm = mcolors.BoundaryNorm(bounds, cmap.N)

    # Plot the board
    plt.figure(figsize=(10, 10))
    plt.imshow(board, cmap=cmap, norm=norm, origin="upper")

    # Draw grid lines
    for i in range(BOARD_ROWS + 1):
        plt.axhline(i - 0.5, color='black', linewidth=0.5)
    for j in range(BOARD_COLS + 1):
        plt.axvline(j - 0.5, color='black', linewidth=0.5)

    plt.xticks(range(BOARD_COLS))
    plt.yticks(range(BOARD_ROWS))
    plt.grid(False)  # Turn off default Matplotlib grid to avoid overlap
    plt.show()

# Backtracking algorithm to solve the puzzle
def solve_puzzle(board, piece_paths, piece_ids, index=0):
    if index == len(piece_ids):
        # All pieces have been placed successfully
        print("Solution found:")
        draw_board(board)
        return True

    piece_id = piece_ids[index]
    paths = piece_paths[piece_id]

    # Try all paths (rotations) for the current piece
    for path in paths:
        coords = path_to_coords(path)
        print(f"Trying piece {piece_id}, path: {path}")

        # Iterate over all possible positions on the board
        for x in range(board.shape[0]):
            for y in range(board.shape[1]):
                # Check if the piece can be placed at (x, y)
                if can_place_piece(board, coords, x, y):
                    print(f"Piece {piece_id} fits at ({x}, {y})")
                    #visualize_piece(board, coords, x, y, piece_id)

                    # Place the piece on the board
                    new_board = place_piece(board, coords, x, y, piece_id)

                    # Recursively try to place the next piece
                    if solve_puzzle(new_board, piece_paths, piece_ids, index + 1):
                        return True  # Solution found

                    # If no solution, backtrack by removing the piece
                    print(f"Backtracking piece {piece_id} at ({x}, {y})")
                    new_board = remove_piece(board, coords, x, y)
                #else:
                    #print(f"Piece {piece_id} does NOT fit at ({x}, {y})")

    # If no placement works for this piece, backtrack
    print(f"No valid placement for piece {piece_id}")
    return False

# Solve the puzzle with the forced order [4, 1, 3, 2]
piece_ids = [4, 1, 3, 2]  # Forced order
if solve_puzzle(board, piece_paths, piece_ids):
    print("Puzzle solved!")
else:
    print("No solution found.")
```
