```python
import numpy as np

def rotate_piece(piece, rotation):
    """Rotate the piece 90 degrees 'rotation' times."""
    return np.rot90(piece, rotation)

def flip_piece(piece):
    """Flip the piece horizontally."""
    return np.fliplr(piece)

def can_place(board, piece, x, y):
    """Check if the piece can be placed at (x, y) on the board."""
    rows, cols = piece.shape
    if x + rows > BOARD_ROWS or y + cols > BOARD_COLS:
        return False  # Out of bounds

    for i in range(rows):
        for j in range(cols):
            if piece[i, j] != 0 and board[x + i, y + j] != 0:
                return False  # Overlap with an existing piece or blocked space
    return True

def place_piece(board, piece, x, y, piece_id):
    """Place the piece on the board."""
    new_board = board.copy()
    rows, cols = piece.shape
    for i in range(rows):
        for j in range(cols):
            if piece[i, j] != 0:
                new_board[x + i, y + j] = piece_id
    return new_board

def solve(board, pieces, index=0):
    """Recursive backtracking solver with debugging."""
    if index == len(pieces):
        return board  # All pieces placed

    piece_id = index + 1
    piece = pieces[index]

    for rotation in range(4):  # Try 4 rotations
        rotated_piece = rotate_piece(piece, rotation)
        flipped_piece = flip_piece(rotated_piece)

        for p in [rotated_piece, flipped_piece]:  # Try original and flipped
            for x in range(BOARD_ROWS):
                for y in range(BOARD_COLS):
                    if can_place(board, p, x, y):
                        print(f"Placing piece {piece_id} at ({x}, {y})")
                        new_board = place_piece(board, p, x, y, piece_id)
                        result = solve(new_board, pieces, index + 1)
                        if result is not None:
                            return result  # Solution found
                        print(f"Backtracking piece {piece_id} from ({x}, {y})")
    print(f"No valid placement found for piece {piece_id}")
    return None  # No valid placement found

# Solve the puzzle
solved_board = solve(board, pieces)
print("Final Board:")
print(solved_board if solved_board is not None else "No solution found")
```
